```python
import pandas as pd
df = pd.read_excel(r"C:\Users\hp\Downloads\ola_project.xlsx")
df.head()
```

```python
num_cols = ['C_TAT','V_TAT', 'Driver_Ratings', 'Customer_Rating']

for col in num_cols:
    df[col] = df.groupby('Vehicle_Type')[col] \
                .transform(lambda x: x.fillna(x.median()))
```

```python
df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna('No')

df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna('No')

df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna('No')

df['Incomplete_Rides_Reason'] = df['Incomplete_Rides_Reason'].fillna('Not Applicable')

df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.head()
```

```python
df=df.drop("vehicle_images", axis =1)
df.head()
```

```python
bins = [0, 5, 15, 30, 50]

labels = ['Short', 'Medium', 'Long', 'Very Long']

df['distance_type'] = pd.cut(
    df['ride_distance'],
    bins=bins,
    labels=labels
)
df.head()
```
